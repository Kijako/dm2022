```python
# Let's transform the [[row][columns][R,G,B]] array into a [row*columns][R,G,B,X,Y] array
# we create an "empty" array with the good dimensions
image_to_list = np.zeros((row_size * col_size, 5), dtype=int)

# we fill it with the good values
for i in range(row_size):
    for j in range(col_size):

        # We get the current index/count of the for loop
        compteur = i * col_size + j

        # we assign the RGBXY values to this index of the new array
        image_to_list[compteur][0] = data[i][j][0]
        image_to_list[compteur][1] = data[i][j][1]
        image_to_list[compteur][2] = data[i][j][2]
        image_to_list[compteur][3] = i
        image_to_list[compteur][4] = j
```
